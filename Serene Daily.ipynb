```python
import os
from datetime import datetime

def create_journal_entry():
    # Folder to store journal entries
    folder_name = "my_journal"
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    
    # Get current date and time
    now = datetime.now()
    date_str = now.strftime("%Y-%m-%d")
    timestamp = now.strftime("%H:%M:%S")
    
    # Filename based on date
    filename = f"{folder_name}/{date_str}.txt"
    
    # Get input from user
    print(f"--- Journal Entry for {date_str} {timestamp} ---")
    entry = input("What's on your mind?\n")
    
    # Write to file
    with open(filename, "a") as file:
        file.write(f"--- {timestamp} ---\n")
        file.write(entry + "\n\n")

    print()
    print("Entry saved!")
    print()
```

```python
def read_entry(entry_name):
    filepath = f"my_journal/{entry_name}"
    try:
        with open(filepath, "r") as file:
            print(file.read())
    except FileNotFoundError:
        print("Entry not found.")
```

```python
def view_entries():
    entries = os.listdir("my_journal")
    if entries:
        for entry in entries:
            print(entry)
        else:
            print("No entries found.")
```

```python
print("\n===== Welcome to your Mental Health Journal! =====")
print("1. Write new entry")
print("2. Read a specific entry")
print("3. View all entries")
print("4. Exit")
print()

while True:  
    choice = input("Choose an option: ")
    print()
    
    if choice == "1":
        create_journal_entry()
    elif choice == "2":
        read_entry()
    elif choice == "3":
        view_entries()
    elif choice == "4":
        print("👋 Goodbye! Keep writing daily.")
        break
    else:
        print("❌ Invalid choice, try again.")
```
